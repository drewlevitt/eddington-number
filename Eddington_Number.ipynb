```python
import pandas as pd
```

```python
df = pd.read_json('DSL_activities_2021-12-27.json')
df
```

```python
df.columns
```

```python
df['date'] = df['start_date_local'].str[:10]
df
```

```python
days = df[df['type'] == 'Ride'].groupby('date')['distance'].sum().sort_values(ascending=False)
days /= 1609.34
days
```

```python
days = pd.concat([days, days.rank(method='first', ascending=False)], axis=1)
days.columns = ['miles', 'rank']
days
```

```python
days['delta'] = days['miles'] - days['rank']
days[days['delta'] >= 0]
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.axis('equal')

days[days['delta'] >= 0].plot(x='rank', y='miles', ax=ax)
```

```python
days.to_csv('days_DSL_2021-12-27.csv')
```
